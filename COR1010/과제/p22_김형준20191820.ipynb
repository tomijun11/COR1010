```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')

df = pd.read_csv('StuData.csv',index_col=0)
print(df)
print()

df['Com'] = [78,99,90,45,80,75]
df['평균'] = df.mean(axis=1,numeric_only=True)
print(df)
print()

plt.title('Subject scores and average for each student')
plt.plot(df['평균'])
plt.plot(df['국어'])
plt.plot(df['영어'])
plt.plot(df['수학'])
plt.plot(df['Com'])
plt.legend(['평균','국어','영어','수학','Com'])
plt.show()

df.loc['평균'] = (df.mean(axis=0,numeric_only=True)).round(2)
print(df)
print()

plt.figure()
plt.title('Average for each subject')
plt.plot(df[['국어','영어','수학','Com']].loc['평균'])
plt.show()
```
