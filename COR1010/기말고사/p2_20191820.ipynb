```python
import pandas as pd

plt.rc('font',family='Malgun Gothic')
df = pd.read_csv('ExamData.csv',index_col=0)

df['Com'] = [78,66,80,45,70,75,70,82,100,85]
df['평균'] = df.mean(axis=1,numeric_only=True).round(2)
df = df.sort_values(by='평균',ascending=False)
df
```

```python
df.plot(kind='barh',title='학생별 과목점수와 평균')
```
